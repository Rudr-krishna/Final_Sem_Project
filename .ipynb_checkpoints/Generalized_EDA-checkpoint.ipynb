```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings
warnings.filterwarnings('ignore')
```

```python
def adf_test(series):
    result = adfuller(series.dropna(), autolag='AIC')
    labels = ['ADF Statistic', 'p-value', '# Lags Used', 'Number of Observations']
    out = pd.Series(result[:4], index=labels)
    for key, val in result[4].items():
        out[f'Critical Value ({key})'] = val
    return out
```

```python
def perform_stock_eda(ticker):
    print(f"\n{'='*30}\nEDA for {ticker}\n{'='*30}")
    
    # Load and preprocess data
    path = f"C:/Users/Shush/Final_Sem_Project/Stocks_Data/{ticker}_data.csv"
    df = pd.read_csv(path)
    
    df.columns = df.iloc[0]
    df = df[3:].reset_index(drop=True)
    df.columns = ["Date", "Close", "High", "Low", "Open", "Volume"]
    df.insert(1, "Ticker", ticker)
    
    df["Open"] = df["Open"].astype(float)
    df["High"] = df["High"].astype(float)
    df["Low"] = df["Low"].astype(float)
    df["Close"] = df["Close"].astype(float)
    df["Volume"] = df["Volume"].astype(int)
    
    df['Date'] = pd.to_datetime(df['Date'])
    df.index = df.pop('Date')
    
    # Summary
    print("\nMissing Values:\n", df.isnull().sum())
    print("\nData Info:")
    print(df.info())
    print("\nDescriptive Stats:")
    print(df.describe())

    # Plot Close Price
    plt.figure(figsize=(14, 7))
    plt.plot(df.index, df['Close'], label='Close Price', color='blue')
    plt.xticks(rotation=45)
    plt.title(f"{ticker} Stock Price Over Time")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.show()

    # Moving Averages
    df['MA7'] = df['Close'].rolling(window=7).mean()
    df['MA30'] = df['Close'].rolling(window=30).mean()

    plt.figure(figsize=(14, 7))
    plt.plot(df['Close'], label='Close Price', color='blue', alpha=0.6)
    plt.plot(df['MA7'], label='7-day Moving Average', color='red', linestyle="dashed")
    plt.plot(df['MA30'], label='30-day Moving Average', color='green', linestyle="dashed")
    plt.title(f"{ticker} Moving Averages")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.show()

    # ADF Test
    print("\nADF Test on Closing Prices:")
    print(adf_test(df['Close']))

    # ACF and PACF plots
    plt.figure(figsize=(14, 5))
    plot_acf(df['Close'].dropna(), lags=40)
    plt.title(f"{ticker} - Autocorrelation")
    plt.show()

    plt.figure(figsize=(14, 5))
    plot_pacf(df['Close'].dropna(), lags=40)
    plt.title(f"{ticker} - Partial Autocorrelation")
    plt.show()
```

```python
# List of tickers
tickers = ["MSFT", "AAPL", "NVDA", "AMZN", "TSLA"]

# Running EDA for each company
for ticker in tickers:
    perform_stock_eda(ticker)
```
