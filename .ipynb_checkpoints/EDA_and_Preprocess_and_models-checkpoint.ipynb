```python
def adf_test(series):
    result = adfuller(series.dropna(), autolag='AIC')
    labels = ['ADF Statistic', 'p-value', '# Lags Used', 'Number of Observations']
    out = pd.Series(result[:4], index=labels)
    for key, val in result[4].items():
        out[f'Critical Value ({key})'] = val
    return out

# Test on Closing Price
print(adf_test(df['Close']))
```

# Preprocessing

```python
df_arima = df[['Close']].copy()
df_arima['Close_diff'] = df_arima['Close'].diff().dropna()  # Differencing
df_arima.dropna(inplace=True)
```

```python
df_arimax = df[['Close', 'MA7', 'MA30', 'Volume']].copy()
df_arimax.dropna(inplace=True)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df_lstm = df[['Close']].copy()
df_lstm['Scaled_Close'] = scaler.fit_transform(df_lstm[['Close']])

# Creating time-series sequences (lookback = 60 days)
lookback = 60
X, y = [], []
for i in range(lookback, len(df_lstm)):
    X.append(df_lstm['Scaled_Close'].iloc[i-lookback:i].values)
    y.append(df_lstm['Scaled_Close'].iloc[i])

X, y = np.array(X), np.array(y)

# Reshaping for LSTM input (samples, timesteps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

print("LSTM Input Shape:", X.shape)
```

```python
train_size = int(len(df) * 0.8)

# Splitting for ARIMA
train_arima, test_arima = df_arima[:train_size], df_arima[train_size:]

# Splitting for ARIMAX
train_arimax, test_arimax = df_arimax[:train_size], df_arimax[train_size:]

# Splitting for LSTM
X_train, X_test, y_train, y_test = X[:train_size-lookback], X[train_size-lookback:], y[:train_size-lookback], y[train_size-lookback:]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error

# Define ARIMA model (using p=1, d=1, q=1 as an initial guess)
model_arima = ARIMA(train_arima['Close'], order=(1, 1, 1))
model_arima = model_arima.fit()

# Make predictions
pred_arima = model_arima.forecast(steps=len(test_arima))
```

```python
# Evaluate performance
mae_arima = mean_absolute_error(test_arima['Close'], pred_arima)
print(f"ARIMA MAE: {mae_arima}")
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define ARIMAX model with exogenous features (Volume, MA7, MA30)
model_arimax = SARIMAX(train_arimax['Close'], exog=train_arimax[['Volume', 'MA7', 'MA30']], order=(1, 1, 1))
model_arimax = model_arimax.fit()

# Make predictions
pred_arimax = model_arimax.forecast(steps=len(test_arimax), exog=test_arimax[['Volume', 'MA7', 'MA30']])
```

```python
mae_arimax = mean_absolute_error(test_arimax['Close'], pred_arimax)
print(f"ARIMAX MAE: {mae_arimax}")
```

```python
# Define LSTM Model
model_lstm = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(units=50, return_sequences=True),
    Dropout(0.2),
    LSTM(units=50),
    Dropout(0.2),
    Dense(units=1)  # Output Layer (1 neuron for price prediction)
])

# Compile Model
model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Early Stopping (to avoid overfitting)
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train Model
history = model_lstm.fit(
    X_train, y_train,
    epochs=10, batch_size=16,
    validation_data=(X_test, y_test),
    callbacks=[early_stop],
    verbose=1
)

# Predict Stock Prices
pred_lstm = model_lstm.predict(X_test)

# Inverse Transform Predictions
pred_lstm = scaler.inverse_transform(pred_lstm.reshape(-1, 1))
```

```python
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error
from itertools import product
```

```python
df_arimax_opt = df.copy()

df_arimax_opt['Return'] = df_arimax_opt['Close'].pct_change()  # Log Returns
df_arimax_opt['SMA_10'] = df_arimax_opt['Close'].rolling(window=10).mean()  # 10-day Simple Moving Average
df_arimax_opt['SMA_30'] = df_arimax_opt['Close'].rolling(window=30).mean()  # 30-day Simple Moving Average
df_arimax_opt.dropna(inplace=True)  # Drop NaN values
```

```python
# STEP 3: Train-Test Split
train_size = int(len(df_arimax_opt) * 0.8)
train, test = df_arimax_opt.iloc[:train_size], df_arimax_opt.iloc[train_size:]

# Define Target Variable (Close Price) and Exogenous Variables
X_train, y_train = train[['Return', 'SMA_10', 'SMA_30']], train['Close']
X_test, y_test = test[['Return', 'SMA_10', 'SMA_30']], test['Close']
```

```python
p_values = range(0, 3)
d_values = range(0, 2)
q_values = range(0, 3)
best_mae = float('inf')
best_order = None

for p, d, q in product(p_values, d_values, q_values):
    try:
        model = SARIMAX(y_train, exog=X_train, order=(p, d, q), seasonal_order=(0, 0, 0, 0))
        results = model.fit(disp=False)
        pred = results.predict(start=len(train), end=len(df_arimax)-1, exog=X_test)
        mae = mean_absolute_error(y_test, pred)

        if mae < best_mae:
            best_mae = mae
            best_order = (p, d, q)
            best_model = results
    except:
        continue

print(f"Best ARIMAX Order: {best_order}, Best MAE: {best_mae}")
```

```python
final_model = SARIMAX(y_train, exog=X_train, order=best_order, seasonal_order=(0, 0, 0, 0))
final_results = final_model.fit(disp=False)
pred_arimax = final_results.predict(start=len(train), end=len(df_arimax)-1, exog=X_test)

# 📌 STEP 6: Plot Predictions vs Actual Prices
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(test.index, y_test, label="Actual Prices", color="green")
plt.plot(test.index, pred_arimax, label="ARIMAX Prediction", color="red")
plt.title("ARIMAX Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# 📌 STEP 7: Evaluate Model
mae_arimax = mean_absolute_error(y_test, pred_arimax)
print(f"Optimized ARIMAX MAE: {mae_arimax}")
```

```python
# 📌 STEP 1: Create a Copy for ARIMAX
df_arimax = df.copy()

# 📌 STEP 2: Feature Engineering (Adding More Features)
df_arimax['Return'] = df_arimax['Close'].pct_change()  # Log Returns
df_arimax['SMA_10'] = df_arimax['Close'].rolling(window=10).mean()  # 10-day Simple Moving Average
df_arimax['SMA_30'] = df_arimax['Close'].rolling(window=30).mean()  # 30-day Simple Moving Average
df_arimax['EMA_10'] = df_arimax['Close'].ewm(span=10, adjust=False).mean()  # 10-day Exponential Moving Average
df_arimax['EMA_30'] = df_arimax['Close'].ewm(span=30, adjust=False).mean()  # 30-day Exponential Moving Average
df_arimax['Volatility'] = df_arimax['Return'].rolling(window=10).std()  # Rolling Volatility
df_arimax['Close_diff'] = df_arimax['Close'].diff()  # Differencing for stationarity

df_arimax.dropna(inplace=True)  # Drop NaN values

# 📌 STEP 3: Train-Test Split
train_size = int(len(df_arimax) * 0.8)
train, test = df_arimax.iloc[:train_size], df_arimax.iloc[train_size:]

# Define Target Variable (Close Price) and Exogenous Variables
X_train, y_train = train[['Return', 'SMA_10', 'SMA_30', 'EMA_10', 'EMA_30', 'Volatility']], train['Close_diff']
X_test, y_test = test[['Return', 'SMA_10', 'SMA_30', 'EMA_10', 'EMA_30', 'Volatility']], test['Close_diff']

# 📌 STEP 4: ARIMAX Hyperparameter Tuning
p_values = range(0, 3)
d_values = range(0, 2)
q_values = range(0, 3)
best_mae = float('inf')
best_order = None

for p, d, q in product(p_values, d_values, q_values):
    try:
        model = SARIMAX(y_train, exog=X_train, order=(p, d, q), seasonal_order=(0, 0, 0, 0))
        results = model.fit(disp=False)
        pred = results.predict(start=len(train), end=len(df_arimax)-1, exog=X_test)
        mae = mean_absolute_error(y_test, pred)

        if mae < best_mae:
            best_mae = mae
            best_order = (p, d, q)
            best_model = results
    except:
        continue

print(f"Best ARIMAX Order: {best_order}, Best MAE: {best_mae}")
```

```python
final_model = SARIMAX(y_train, exog=X_train, order=best_order, seasonal_order=(0, 0, 0, 0))
final_results = final_model.fit(disp=False)
pred_arimax = final_results.predict(start=len(train), end=len(df_arimax)-1, exog=X_test)

# 📌 STEP 6: Convert Predictions Back (Undo Differencing)
pred_arimax = test['Close'].iloc[0] + np.cumsum(pred_arimax)

# 📌 STEP 7: Plot Predictions vs Actual Prices
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(test.index, test['Close'], label="Actual Prices", color="green")
plt.plot(test.index, pred_arimax, label="Improved ARIMAX Prediction", color="red")
plt.title("Enhanced ARIMAX Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# 📌 STEP 8: Evaluate Model
mae_arimax = mean_absolute_error(test['Close'], pred_arimax)
print(f"Optimized ARIMAX MAE: {mae_arimax}")
```
